```python
work_path = "/content/drive/MyDrive/PaperRecommendation/"
checkpoint_path = work_path + "checkpoint/"
```

```python
data_train = pd.read_csv(checkpoint_path + "processed_data/train.csv", encoding = "ISO-8859-1")
data_train = data_train.sample(n=1000, random_state=42)

data_validate = pd.read_csv(checkpoint_path + "processed_data/validate.csv", encoding = "ISO-8859-1")
data_validate = data_validate.sample(n=1000, random_state=42)

data_test = pd.read_csv(checkpoint_path + "processed_data/test.csv", encoding = "ISO-8859-1")
data_test = data_test.sample(n=1000, random_state=42)

data_aims = pd.read_csv(checkpoint_path + "processed_data/aims.csv", encoding = "ISO-8859-1")

data_train.to_csv(checkpoint_path + "sub_processed_data/train.csv", index=False)
data_validate.to_csv(checkpoint_path + "sub_processed_data/validate.csv", index=False)
data_test.to_csv(checkpoint_path + "sub_processed_data/test.csv", index=False)
data_aims.to_csv(checkpoint_path + "sub_processed_data/aims.csv", index=False)
```
